#Challenge Telecom X - Parte 2

```python
import pandas as pd
ruta= '/content/datos_tratados.csv'
datos= pd.read_csv(ruta)
datos
```

```python
#eliminar columna cutomerID
datos= datos.drop('customerID', axis=1)
```

```python
datos.info()
```

```python
datos.columns
```

```python
#ver datos únicos de cada columna
columnas= datos.columns
for columna in columnas:
  print(columna)
  print(datos[columna].unique())
```

```python
categoricas=['gender','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaymentMethod']
```

```python
#codificar columnas categóricas
datos_codificados= pd.get_dummies(data=datos, columns=categoricas, dtype=int)
datos_codificados.sample(5)
```

```python
datos.value_counts(subset='Churn', normalize=True)
```

```python
datos_codificados.columns
```

```python
#separar x e y
X = datos_codificados.drop('Churn', axis=1)
y = datos_codificados['Churn']
```

```python
#dividir en conjunto de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)
```

```python
y_train.value_counts()
```

```python
#oversampling
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=7)
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)
```

```python
y_train_over.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
arbol = DecisionTreeClassifier(max_depth = 4)
random_forest = RandomForestClassifier(max_depth = 3)
```

```python
arbol.fit(X_train_over, y_train_over)
print(f'Exactitud del modelo con el conjunto de entrenamiento es: {arbol.score(X_train_over, y_train_over)}')
print(f'Exactitud del modelo con el conjunto de prueba es: {arbol.score(X_test, y_test)}')
```

```python
random_forest.fit(X_train_over, y_train_over)
print(f'Exactitud del modelo con el conjunto de entrenamiento es: {random_forest.score(X_train_over, y_train_over)}')
print(f'Exactitud del modelo con el conjunto de validación es: {random_forest.score(X_test, y_test)}')
```
